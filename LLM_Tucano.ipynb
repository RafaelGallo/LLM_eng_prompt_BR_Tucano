```python
import requests
import matplotlib.pyplot as plt

import torch
from transformers import AutoModelForCausalLM, AutoProcessor
from transformers import AutoTokenizer, AutoModelForCausalLM

from PIL import Image
from io import BytesIO
```

```python
import torch
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
```

## **Modelo 1 - Tucano-1b1**

```python
# Nome do modelo
model_name = "TucanoBR/Tucano-1b1"
```

```python
# Verificar se a GPU está disponível
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Usando o dispositivo: {device}")
```

```python
# Carregar o tokenizador e o modelo pré-treinado
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Mover o modelo para a GPU (se disponível)
model = model.to(device)
```

```python
# Criar uma pipeline de geração de texto
pipe = pipeline("text-generation", model=model,
                tokenizer=tokenizer,
                device=0 if torch.cuda.is_available() else -1)
```

```python
# Definir o prompt
prompt = (
    "Explique por que a preservação dos biomas brasileiros, como o Cerrado e a Caatinga, "
    "é essencial para a biodiversidade e a segurança hídrica no Brasil."
)
```

```python
# Gerar o texto
result = pipe(prompt, max_length=150, num_return_sequences=1, do_sample=True)

# Exibir o resultado gerado
print("Texto gerado:")
print(result[0]["generated_text"])
```

# **Eng prompt**

```python
import torch
import requests
import matplotlib.pyplot as plt
from io import BytesIO
from transformers import AutoTokenizer, AutoModelForCausalLM
from PIL import Image
```

```python
# Nome do modelo
model_name2 = "TucanoBR/Tucano-630m"

# Verificar se a GPU está disponível
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Usando o dispositivo: {device}")
```

```python
# Carregar o tokenizador e o modelo pré-treinado
tokenizer = AutoTokenizer.from_pretrained(model_name2)
model = AutoModelForCausalLM.from_pretrained(model_name2)

# Mover o modelo para a GPU (se disponível)
model = model.to(device)
```

```python
# URL da imagem
image_file = "https://www.holmy.com.br/wp-content/uploads/2021/10/IMG_1646-1200x900.jpeg"

# Baixando e carregando a imagem
response = requests.get(image_file)
image = Image.open(BytesIO(response.content))

# Exibindo a imagem
plt.imshow(image)
plt.axis('off')  # Remove os eixos
plt.show()
```

```python
# Prompt descrevendo a imagem
prompt = ("Quais os principais elementos desta imagem: "
          "Uma casa ampla com arquitetura moderna localizada em um campo verde, cercada por árvores altas e um lago refletindo o cenário ao redor."
          "Há um carro preto estacionado ao lado da casa e um céu limpo ao fundo.")
```

```python
# Tokenizar e gerar resposta
inputs = tokenizer(prompt, return_tensors="pt").to(device)
outputs = model.generate(inputs["input_ids"], max_length=150, num_return_sequences=1, do_sample=True)
output_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
# Imprimir o output gerado
print("Resposta gerada pelo modelo:")
print(output_text)
```
